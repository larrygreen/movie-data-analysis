```python
import pandas as pd
import numpy as np
original = pd.read_csv('/Users/mac/Desktop/Python Projects/IMDB_Ratings.csv')
new_column_names = {'imdbAverageRating': 'imdbRating', 'genre1': 'genre'}
original.rename(columns=new_column_names, inplace=True)
original_corrected = original.replace(
    {'What&#39;s Eating Gilbert Grape': "What's Eating Gilbert Grape", 
     'Devil&#39;s Advocate, The': "The Devil's Advocate", 'Thelma &amp; Louise': 'Thelma and Louise',
     'Harry Potter and the Sorcerer&#39;s Stone': "Harry Potter and the Sorcerer's Stone",
     'I&#39;ve Loved You So Long': "I've Loved You So Long", 
     'King of Devil&#39;s Island': "King of Devil's Island", 
     'Fanny, Annie &amp; Danny':'Fanny, Annie and Danny', 
     'Dolly Parton&#39;s Coat of Many Colors': "Dolly Parton's Coat of Many Colors",
     'God&#39;s Own Country': "God's Own Country"})
movies = original_corrected[['title', 'year', 'imdbRating', 'genre']]
original_shape = original.shape
print(f"Original DataFrame Rows and Columns: {original_shape}")
```

```python
unique_genres = movies['genre'].unique()
print(unique_genres)
```

```python
x = movies[(movies['genre'] == 'Action') & 
    (movies['imdbRating'] >= 7.5) & (movies['year'] >= 1990) & 
    (movies['year'] <= 1999)]
high_quality_90s_action = x.drop([2785, 2893, 3035, 3287], axis = 0)
high_quality_90s_action
```

```python
high_quality_90s_adventure = movies[(movies['genre'] == 'Adventure') & 
    (movies['imdbRating'] >= 7.5) & (movies['year'] >= 1990) & 
    (movies['year'] <= 1999)]
high_quality_90s_adventure
```

```python
y = movies[(movies['genre'] == 'Comedy') & 
    (movies['imdbRating'] >= 7.5) & (movies['year'] >= 1990) & 
    (movies['year'] <= 1999)]
high_quality_90s_comedy = y.drop([3190, 3290], axis = 0)
high_quality_90s_comedy
```

```python
pd.set_option('display.max_rows', None)
z = movies[(movies['genre'] == 'Drama') & 
    (movies['imdbRating'] >= 7.5) & (movies['year'] >= 1990) & 
    (movies['year'] <= 1999)]
high_quality_90s_drama = z.drop([3230, 3464], axis = 0)
high_quality_90s_drama
```

```python
combined_90s = pd.concat([high_quality_90s_action, high_quality_90s_adventure, 
                high_quality_90s_comedy, high_quality_90s_drama])
combined_90s
```

```python
high_quality_each_year = combined_90s.groupby('year', as_index=False).agg(
    {'title':'count'})
high_quality_each_year = high_quality_each_year.rename(
    columns={'title': 'high quality movies'})
high_quality_each_year
```

```python
high_quality_by_genre_90s = combined_90s.groupby('genre', as_index=False).agg(
    {'title':'count'})
high_quality_by_genre_90s = high_quality_by_genre_90s.rename(
    columns={'title': 'high quality movies'})
high_quality_by_genre_90s
```
